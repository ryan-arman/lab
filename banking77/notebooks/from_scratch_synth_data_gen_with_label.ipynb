```python
import subprocess
import os
from pathlib import Path

# Base paths
script_path = "/Users/ryanarman/code/lab/banking77/notebooks/convert_format.py"
data_dir = "/Users/ryanarman/code/lab/banking77/notebooks/data/synth"

# Input files (with _with_label suffix)
val_input = f"{data_dir}/banking77_synth_val_with_label.jsonl"
test_input = f"{data_dir}/banking77_synth_test_with_label.jsonl"

# System prompt options
prompt_types = ['full', 'basic', 'synth', 'empty']

# Run conversion for each prompt type
for prompt_type in prompt_types:
    print(f"\n{'='*60}")
    print(f"Processing with system prompt: {prompt_type}")
    print(f"{'='*60}")
    
    # Generate output filenames with suffix
    val_output = f"{data_dir}/banking77_synth_val_{prompt_type}.jsonl"
    test_output = f"{data_dir}/banking77_synth_test_{prompt_type}.jsonl"
    
    # Convert validation set
    print(f"\nConverting validation set...")
    cmd_val = [
        "python", script_path,
        val_input,
        val_output,
        "--reverse",
        "--system-prompt", prompt_type
    ]
    result_val = subprocess.run(cmd_val, capture_output=True, text=True)
    if result_val.returncode == 0:
        print(f"✓ Successfully created: {val_output}")
    else:
        print(f"✗ Error converting validation set:")
        print(result_val.stderr)
    
    # Convert test set
    print(f"\nConverting test set...")
    cmd_test = [
        "python", script_path,
        test_input,
        test_output,
        "--reverse",
        "--system-prompt", prompt_type
    ]
    result_test = subprocess.run(cmd_test, capture_output=True, text=True)
    if result_test.returncode == 0:
        print(f"✓ Successfully created: {test_output}")
    else:
        print(f"✗ Error converting test set:")
        print(result_test.stderr)

print(f"\n{'='*60}")
print("All conversions complete!")
print(f"{'='*60}")
```
